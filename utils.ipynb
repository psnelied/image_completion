```python
# mean pixel value of the training dataset
def mean_pixel_value(loader, num):
    mpv = 0
    n_images = 0
    for i, (img,_) in enumerate(loader, 0):
        if i < num : 
            for j in range(len(img)):
                mpv += torch.mean(img[j])
                n_images+=1
        else : break
    return mpv / n_images


def generate_mask(input_shape, patch = (96,96), size = ([48,96], [48,96]), n_holes=1):
    mask = torch.zeros(input_shape)
    batch_size, _, mask_h, mask_w = input_shape

    size_w_min = size[0][0]
    size_w_max = size[0][1]
    size_h_min = size[1][0]
    size_h_max = size[1][1]

    region = 0
    if n_holes==1:
        x_min = random.randint(0, mask_w - patch[0])
        y_min = random.randint(0, mask_h - patch[1])

    for i in range(batch_size):
        for j in range(n_holes):

            if n_holes > 1 : 
                x_min = random.randint(0, mask_w - patch[0])
                y_min = random.randint(0, mask_h - patch[1])

            size_w = random.randint(size_w_min, size_w_max)
            size_h = random.randint(size_h_min, size_h_max)

            x_start = random.randint(x_min, x_min + patch[0] - size_w)
            y_start = random.randint(y_min, y_min + patch[1] - size_h)

            mask[i, 0, y_start : y_start+size_h, x_start : x_start+size_w] = 1

    if n_holes == 1:
        region = ([y_min, y_min + patch[1]],[x_min, x_min + patch[0]])

    return mask, region


def imshow_2(img, show=True): # function from TDs
    img = img.numpy()
    if img.shape[0] == 1:
        img = img/(np.maximum(np.max(img), 0.00001))
        img = np.repeat(img, repeats=3, axis=0)
    plt.imshow(np.transpose(img, (1, 2, 0)))
    if not show :
        plt.axis('off')
    plt.show()


def crop(images, patch):
    x_start, x_end = patch[1]
    y_start, y_end = patch[0]
    return images[:, :, y_start : y_end, x_start : x_end]


#def random_patch(patch_size = (96,96), image_size = (128,192)): #for cars dataset
def random_patch(patch_size = (96,96), image_size = (160,160)):
    patch_size_h = patch_size[0]
    patch_size_w = patch_size[1]
    size_h = image_size[0]
    size_w = image_size[1]
    x_start = random.randint(0, size_w-patch_size_w)
    y_start = random.randint(0, size_h-patch_size_h)
    return ([y_start, y_start+patch_size_h], [x_start, x_start+patch_size_w])
```
